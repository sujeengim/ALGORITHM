### 정렬
- 데이터 집합을 일정한 순서로 바꾸어 늘어놓는 작업
- 안정적인 정렬 알고리즘 stable sorting algorithm
    - 값이 같은 원소의 순서가 정렬한 후에도 유지되는
정렬 알고리즘
- 내부 정렬 internal sorting
    - 정렬할 모든 데이터를 하나의 배열에 저장할 수 있는 경우
    - 처리 위치 : 메인 메모리 (RAM) 내에서 정렬
    - 데이터 크기 : 전체 데이터를 한 번에 메모리에 올릴 수 있어야 함
- 외부 정렬 external sorting
    - 데이터가 너무 커서 메인 메모리에 다 올릴 수 없을 때, 디스크 등 보조 저장장치와 메모리를 병행해서 정렬
    - 처리 위치 : 디스크와 메모리를 병행
    - 데이터 크기 : 메모리보다 훨씬 큰 데이터도 처리 가능

# 버블 정렬
## 문제 예시-실습 6-2) 1,3,5,7,9 를 버블 솔팅하는 과정을 보이시오.
- 이웃한 두 원소의 대소 관계를 비교하여 필요에 따라
교환을 반복하는 알고리즘
- 단순 교환 정렬

- 패스 pass
    - 일련의 비교·교환하는 과정
    - 패스를 한 번 수행할 때마다 정렬할 대상이 1개씩 감소

image.png

```python
# 버블 정렬 구현
from typing import MutableSequence

def bubble_sort(a: MutableSequence) -> None:
    n = len(a)
    for i in range(n-1):
        for j in range(n-1, i, -1):
            if a[j-1] > a[j]:
                a[j-1], a[j] = a[j], a[j-1]

if __name__ == '__main__':
    print('버블 정렬을 수행합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))

    bubble_sort(x)

    print('오름 차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
```

```python
# 교환 과정 출력
from typing import MutableSequence

def bubble_sort_verbose(a: MutableSequence) -> None:
    ccnt = 0
    scnt = 0
    n = len(a)
    for i in range(n - 1):
        print(f'<pass {i + 1}>')
        for j in range(n - 1, i, -1):
            for j in range(n-1, i, -1):
                for m in range(n-1):
                    print(f'{a[m]:2}'+ (' 'if m != j-1 else
                                        ' +' if a[j-1]>a[j] else ' -'),
                          end='')
                print(f'{a[n-1]:2}')
                ccnt += 1
                if a[j-1] > a[j]:
                    scnt += 1
                    a[j-1], a[j] = a[j], a[j-1]
            for m in range(n-1):
                print(f'{a[m]:2}', end=' ')
            print(f'{a[n-1]:2}')
        print(f'비교를 {ccnt}번 했습니다.')
        print(f'교환을 {scnt}번 했습니다.')

a = [6,4,3,7,1,9,8]
bubble_sort_verbose(a)
```

```python
# 버블 정렬 개선(1)
def bubble_sort2_verbose(a: MutableSequence) -> None:
    n = len(a)
    ccnt = 0
    scnt = 0
    for i in range(n - 1):
        print(f'<pass {i + 1}>')
        exchng = 0
        for j in range(n - 1, i, -1):
            for m in range(n-1):
                print(f'{a[m]:2}' + (' ' if m != j-1 else
                                     ' +' if a[j-1] > a[j] else ' -'),
                      end = '')
            print(f'{a[n-1]:2}')
            ccnt += 1
            if a[j-1] > a[j]:
                a[j-1], a[j] = a[j], a[j-1]
                scnt += 1
                exchng += 1
        for m in range(n-1):
            print(f'{a[m]:2}', end = ' ')
        print(f'{a[n-1]:2}')
        if exchng == 0:
            break
    print(f'비교를 {ccnt}번 했습니다.')
    print(f'교환을 {scnt}번 했습니다.')

bubble_sort2_verbose([6,4,3,7,1,9,8])
```

```python
# 버블 개선 2
def bubble_sort3_verbose(a: MutableSequence) -> None:
    n = len(a)
    k = 0
    i = 0
    ccnt = 0
    scnt = 0
    while k<n-1:
        print(f'<pass {i + 1}>')
        last = n-1
        for j in range(n-1, k, -1):
            for m in range(n-1):
                print(f'{a[m]:2}' + (' ' if m != j-1 else
                                     ' +' if a[j-1] > a[j] else ' -'),
                      end = '')
            print(f'{a[n-1]:2}')
            ccnt += 1
            if a[j-1] > a[j]:
                a[j-1], a[j] = a[j], a[j-1]
                scnt += 1
                last = j
        k = last
        i += 1
        for m in range(n-1):
            print(f'{a[m]:2}', end = ' ')
        print(f'{a[n-1]:2}')
    print(f'비교를 {ccnt}번 했습니다.')
    print(f'교환을 {scnt}번 했습니다.')

bubble_sort3_verbose([6,4,3,7,1,9,8])
```

### 셰이커 정렬
- 정렬이 거의 된 경우 버블정렬 개선 방법
- 홀수패스에서가장작은원소를맨앞으로이동
- 짝수패스에서가장큰원소를맨뒤로이동
- 스캔범위첫원소인덱스: left
- 스캔범위마지막원소인덱스: right

```python
# 셰이커 정렬
def shaker_sort_verbose(a: MutableSequence) -> None:
    n = len(a)
    left = 0
    right = n-1
    last = right
    ccnt = 0
    scnt = 0
    i = 0
    while left < right:
        print(f'<pass {i+1}>')
        for j in range(right, left, -1):
            for m in range(n-1):
                print(f'{a[m]:2}' + (' ' if m != j-1 else
                                     ' +' if a[j-1] > a[j] else ' -'),
                      end = '')
            print(f'{a[n-1]:2}')
            ccnt += 1
            if a[j-1] > a[j]:
                a[j-1], a[j] = a[j], a[j-1]
                scnt += 1
                last = j
        left = last
        for m in range(n-1):
            print(f'{a[m]:2}', end = ' ')
        print(f'{a[n-1]:2}')
        i += 1
        if not left < right:
            break

        print(f'<pass {i+1}>')
        for j in range(left, right):
            for m in range(n-1):
                print(f'{a[m]:2}' + (' ' if m != j-1 else
                                     ' +' if a[j-1]>a[j] else ' -'),
                      end = '')
            print(f'{a[n-1]:2}')
            ccnt += 1
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
                scnt += 1
                last = j
        right = last
        for m in range(n-1):
            print(f'{a[m]:2}', end=' ')
        print(f'{a[n-1]:2}')
        i += 1
    print(f'비교를 {ccnt}번 했습니다.')
    print(f'교환을 {scnt}번 했습니다.')

shaker_sort_verbose([9,1,3,4,6,7,8])
```

### 단순 선택 정렬
- 가장 작은 원소부터 선택해 알맞은 위치로 옮기는
작업을 반복하여 정렬

    ---> 오름차순 정렬이라면 가장 작은 원소를 선택해 가장 왼쪽 원소와 교환

- 안정적이지 않은 정렬 알고리즘

image.png

```python
from typing import MutableSequence

def selection_sort(a: MutableSequence) -> None:
    n = len(a)
    for i in range(n-1):
        min = i
        for j in range(i+1, n):
            if a[j] < a[min]:
                min = j
        a[i], a[min] = a[min], a[i]
```

### 단순 삽입 정렬
- 주목한 원소보다 더 앞쪽에서 알맞은 위치로
삽입하며 정렬하는 알고리즘
- 두 번째인 원소부터 주목
- 하나씩 정렬이 되어감
- 떨어져 있는 원소를 교환하지 않아
안정적인 정렬 알고리즘
- 문제
    - 삽입에 필요한 비교, 교환 비용 커짐 -> 이진 삽입 정렬
    - 삽입할 위치가 멀리 떨어져 있으면 이동 횟수 증가 -> 셸 정렬

image.png

```python
from typing import MutableSequence

def insertion_sort(a: MutableSequence) -> None:
    n = len(a)
    for i in range(1,n):
        j = i
        tmp = a[i]
        while j>0 and a[j-1] > tmp:
            a[j] = a[j-1]
            j -= 1
        a[j] = tmp

if __name__ == '__main__':
    print('단순 삽입 정렬을 수행합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))

    insertion_sort(x)

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
```

### 이진 삽입 정렬
- 이진 검색을 이용하여 원소 삽입 위치 계산에 필요
한 비용을 절약
- bisect.insort(a, x, lo, hi)
    - a가 이미 정렬된 상태를 유지하면서, a[lo]~a[hi]
사이에 x를 삽입


image.png

```python
from typing import MutableSequence

def binary_insertion_sort(a: MutableSequence) -> None:
    n = len(a)
    for i in range(1,n):
        key = a[i]
        pl = 0
        pr = i-1

        while True:
            pc = (pl+pr)//2
            if a[pc] == key:
                break
            elif a[pc] < key:
                pl = pc+1
            else:
                pr = pc -1
            if pl > pr:
                break

        pd = pc+1 if pl <= pr else pr+1

        for j in range(i,pd,-1):
            a[j] = a[j-1]
        a[pd] = key

if __name__ == "__main__":
    print('이진 삽입 정렬을 수행합니다.')
```

```python
# bisect.insort 사용
from typing import MutableSequence
import bisect

def binary_insertion_sort(a: MutableSequence) -> None:
    for i in range(1,len(a)):
        bisect.insort(a,a.pop(i),0,i)
```

### 셸 정렬
- 원소를 그룹으로 나눠 각 그룹별로 정렬을 수행한 뒤 정렬된 그룹을 합치는 작업을 반
복하여 원소의 이동 횟수를 줄이는 방식으로 동작
- 퀵 정렬이 고안되기에 앞서 가장 빠른 정렬 알고리즘
- 구현
    - h 값 선택
        - 각 반복마다 **n // 2** --> 그룹간 섞이지 않는 문제 발생
            - h 값들이 서로 배수 관계(예: h = 8, 4, 2, 1)면, <br>
            같은 위치끼리 계속 비교되기 때문에 일부 데이터가 끝까지 정렬되지 않을 수 있음
        - h 값은 서로 배수가 되지 않아야 원소가 섞일 수 있음
            
            -->  비배수 수열

image.png

```python
from typing import MutableSequence

def shell_sort(a: MutableSequence) -> None:
    n = len(a)
    h = n//2
    while h>0:
        for i in range(h,n):
            j = i-h
            tmp = a[i]
            while j>=0 and a[j]>tmp:
                a[j+h] = a[j]
                j -= h
            a[j+h] = tmp
        h//=2

if __name__ == '__main__':
    print('셀 정렬을 수행합니다.')
    num = int(input('원소 수를 입력하세요. '))
    x = [None] * num

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))

    shell_sort(x)

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
```

```python
# 개선된 셀 정렬
from typing import MutableSequence

def shell_sort(a: MutableSequence) -> None:
    n = len(a)
    h = 1

    while h<n //9:
        h = h*3+1

    while h >0:
        for i in range(h,n):
            j = i-h
            tmp = a[i]
            while j>=0 and a[j]>tmp:
                a[j+h] = a[j]
                j -= h
            a[j+h] = tmp
        h //= 3

if __name__ == '__main__':
    print('셀 정렬을 수행합니다.(h*3+1의 수열 사용)')
    num = int(input('원소 수를 입력하세요.'))
    x = [None] * num

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))

    shell_sort(x)

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
```
